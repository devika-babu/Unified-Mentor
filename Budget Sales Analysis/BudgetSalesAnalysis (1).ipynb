```python
### extracting age from the Birthdate columns of the data
from datetime import datetime 
from dateutil.relativedelta import relativedelta

current_date=datetime.now()
current_date
```

```python
df['Age'] = df['BirthDate'].apply(lambda x: relativedelta(current_date, x).years)
```

```python
### Grouping the age in categories for analysis

def Age(years):
    if years<20:
        return 'Teen'
    elif years >20 and years <40:
        return 'Young'
    elif years >40 and years<60:
        return 'Adult'
    else:
        return 'Old'
x=df.loc[113,'Age']
Age(x)
df['AgeCategory']=df['Age'].apply(Age)
df['AgeCategory']
```

```python
### plot of sales accoring to the age category
Sales_agecategory=df.groupby('AgeCategory')['SalesAmount'].sum()
Sales_agecategory.plot(kind='pie',autopct='%0.0f%%',colors=['Magenta','Red'],labels=['Adult','Old'],shadow=True)
plt.title('Sales According to the Age Category',fontsize=20)
Sales_agecategory
```
